```python
# Load relevant packages
import polars as pl
from kedro.framework.session import KedroSession
from kedro.framework.project import configure_project

# Configure the Kedro project
configure_project("optimuskg")

# Load the catalog
session = KedroSession.create()
context = session.load_context()
catalog = context.catalog
```

```python
catalog
```

```python
# Get df names given the URI in the catalog using regular expressions
silver_nodes = catalog.filter(name_regex="silver.nodes") # get list of nodes in silver layer, you can get the edges by using "silver.edges"
silver_nodes
```

```python
silver_nodes_dfs = [catalog.load(df_name) for df_name in silver_nodes]
```

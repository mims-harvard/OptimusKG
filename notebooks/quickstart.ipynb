## Filtering datasets by layer

You can filter datasets by regular expressions. This is useful for filtering nodes in the same layer.

```python
edge_names = catalog.filter(name_regex=re.compile("^silver.edges"))
sorted(edge_names)
```

```python
node_names = catalog.filter(name_regex=re.compile("^silver.nodes"))
sorted(node_names)
```

You can also search for all the datasets that have a specific biological element, since the biological elements presented in a dataset are present in the dataset name.

```python
# Get all the dataset names that have disease data
disease_dataset_names = catalog.filter(name_regex=re.compile("disease"))
sorted(disease_dataset_names)
```

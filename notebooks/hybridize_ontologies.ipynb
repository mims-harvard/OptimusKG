```python
def analyze_owl_ontology(owl_filepath: str) -> None:
    """
    Analyze any OWL ontology file and print out all node types using owlready2.
    
    Args:
        owl_filepath: Path to the OWL file to analyze
    """
    # Load the ontology
    onto = get_ontology(owl_filepath).load()
    
    # Get all classes (node types)
    node_types = set()
    for cls in onto.classes():
        node_types.add(str(cls))
    
    print(f"\nOntology Node Types from {owl_filepath}:")
    print("====================================")
    for node_type in sorted(node_types):
        print(f"- {node_type}")

# Analyze HPO ontology
analyze_owl_ontology("data/landing/ontology/human_phenotype.owl")
```
